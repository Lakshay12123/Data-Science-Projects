```python
#importing tools and lib 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import kagglehub
import warnings
warnings.filterwarnings('ignore')
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score,calinski_harabasz_score
from mpl_toolkits.mplot3d import Axes3D 
from sklearn.cluster import DBSCAN
from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_excel('C:\\Users\\laksh\\.cache\\kagglehub\\datasets\\yasserh\\customer-segmentation-dataset\\versions\\1\\Online Retail.xlsx')
```

```python
df.dropna(inplace=True)
df.drop(df[df.duplicated()].index,axis=0,inplace=True)
```

```python
df_cleaned = df.drop(['InvoiceNo', 'CustomerID','Description','StockCode'], axis=1)
df_encoded = pd.get_dummies(df_cleaned, columns=['Country'])
df_encoded.drop('InvoiceDate',axis=1,inplace=True)
wrong_data = df_encoded[df_encoded['Quantity']<0].index
df_encoded.drop(index=wrong_data,axis=0,inplace=True)
df_encoded.reset_index(drop=True,inplace=True)
```

```python
df_encoded
```

```python
scaler=StandardScaler()
scaled_df=scaler.fit_transform(df_encoded.select_dtypes(exclude='bool'))
scaled_df=pd.DataFrame(scaled_df,columns=df_encoded.select_dtypes(exclude='bool').columns)
```

```python
scaled_df=pd.concat([scaled_df,df_encoded.select_dtypes(include='bool')],axis=1)
```

```python
out=IsolationForest(contamination=0.1)
scaled_df['out']=out.fit_predict(scaled_df)
df.drop(scaled_df[scaled_df['out']==-1].index,axis=0,inplace=True)
scaled_df.drop(scaled_df[scaled_df['out']==-1].index,axis=0,inplace=True)
scaled_df.reset_index(inplace=True)
scaled_df.drop('index',axis=1,inplace=True)
scaled_df.drop('out',axis=1,inplace=True)
```
