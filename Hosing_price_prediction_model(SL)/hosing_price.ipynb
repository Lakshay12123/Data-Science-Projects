```python
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler=StandardScaler()
scaler.fit(X)
scaled_X_train=scaler.transform(X_train)
scaled_X_test=scaler.transform(X_test)
```

```python
base_linear=LinearRegression()
base_linear.fit(scaled_X_train,y_train)
preds=base_linear.predict(scaled_X_test)
rmse=np.sqrt(mean_squared_error(y_test,preds))
print(rmse)
```

```python
scores=[]
for n in range(9):
    poly=PolynomialFeatures(degree=n)
    poly_X_train=poly.fit_transform(scaled_X_train)
    poly_X_test=poly.transform(scaled_X_test)
    model=LinearRegression()
    model.fit(poly_X_train,y_train)
    preds=model.predict(poly_X_test)
    scores.append(np.sqrt(mean_squared_error(y_test,preds)))
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=range(1, len(scores) + 1), y=scores, marker='o')
plt.xlabel("Polynomial Degree")
plt.ylabel("RMSE")
plt.title("RMSE vs Polynomial Degree")
plt.grid(True)
plt.show()
```

```python
final_poly_mod=PolynomialFeatures(degree=2)
poly_X_train=final_poly_mod.fit_transform(scaled_X_train)
poly_X_test=final_poly_mod.transform(scaled_X_test)
model=LinearRegression()
model.fit(poly_X_train,y_train)
train_preds=model.predict(poly_X_train)
test_preds=model.predict(poly_X_test)

print('training error : ',np.sqrt(mean_squared_error(y_train,train_preds)))
print('testing error : ',np.sqrt(mean_squared_error(y_test,test_preds)))
```

```python
train_size_abs, train_scores, test_scores = learning_curve(model, poly_X_train, y_train, train_sizes=np.linspace(0.01,1,40),cv=5,scoring="neg_root_mean_squared_error")
train_error=-train_scores.mean(axis=1)
test_error=-test_scores.mean(axis=1)
plt.plot(train_size_abs,train_error,'r-+',label='train',linewidth=2)
plt.plot(train_size_abs,test_error,'b-',label='test',linewidth=3)
plt.xlabel("Training set size")
plt.ylabel("RMSE")
plt.grid()
plt.legend(loc="upper right")
```

```python
SGD=SGDRegressor(max_iter=1000, learning_rate='invscaling', eta0=0.01)
SGD.fit(poly_X_train,y_train)
preds=SGD.predict(poly_X_test)
print(np.sqrt(mean_squared_error(y_test,preds)))
```

```python
training_score=[]
testing_score=[]
for n in np.logspace(-5, 3, 100):
    
    ridge_model = Ridge(alpha=n)
    ridge_model.fit(poly_X_train,y_train)
    train_preds=ridge_model.predict(poly_X_train)
    test_preds=ridge_model.predict(poly_X_test)
    
    training_score.append(np.sqrt(mean_squared_error(y_train,train_preds)))
    testing_score.append(np.sqrt(mean_squared_error(y_test,test_preds)))

plt.figure()
plt.plot((np.logspace(-5, 3, 100)),training_score,'r-+' ,label='train_scores',linewidth=2)
plt.plot((np.logspace(-5, 3, 100)),testing_score,'b-',label='test_scores',linewidth=3)
plt.ylabel('RMSE')
plt.xlabel('alpha')
    

```
